```python
#import numpy, pandas and matplotlib

import numpy as nm
import matplotlib.pyplot as mtp
import pandas as pd
```

```python
#connect to google drive data (.csv file)

from google.colab import drive
drive.mount('/content/drive')
```

```python
dataset= pd.read_csv('/content/drive/MyDrive/Datasets/salary_data - salary_data.csv')
```

```python
print(dataset.head())
```

```python
print(dataset)
```

```python
x= dataset.iloc[:, :-1].values
y= dataset.iloc[:, -1].values
```

```python
x
```

```python
y
```

```python
# Splitting the dataset into training and test set.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 1/3, random_state=0)  #test size=ONE THIRD of actual data (30), that is 10 data
```

```python
#Fitting the Simple Linear Regression model to the training dataset
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#Prediction of Test and Training set result
y_test_pred= regressor.predict(x_test)
y_train_pred= regressor.predict(x_train)

#actual and predicted Y of training data
print("Y train -> ", y_train)
print("Y train predicted -> ", y_train_pred)
print(y_train.shape, y_train_pred.shape)

#actual and predicted Y of test data
print("Y test -> ", y_test)
print("Y test predicted -> ", y_test_pred)
print(y_test.shape, y_test_pred.shape)
```

```python
#visualizing the Train set results
mtp.scatter(x_train, y_train, color="green")      #actual training data plot
mtp.plot(x_train, y_train_pred, color="red")       #predicted regression line of train data
mtp.title("Salary vs Experience (Training Dataset)")
mtp.xlabel("Years of Experience")
mtp.ylabel("Salary(In Rupees)")
mtp.show()
```

```python
#visualizing the Test set results
mtp.scatter(x_test, y_test, color="blue")    #actual test data plot
mtp.plot(x_test, y_test_pred, color="red")   #predicted regression line of test data
mtp.title("Salary vs Experience (Test Dataset)")
mtp.xlabel("Years of Experience")
mtp.ylabel("Salary(In Rupees)")
mtp.show()
```

```python
dataset= pd.read_csv('/content/drive/MyDrive/price - price.csv')
```

```python
y_test_pred= regressor.predict(x_test)
y_train_pred= regressor.predict(x_train)

#actual and predicted Y of training data
print("Y train -> ", y_train)
print("Y train predicted -> ", y_train_pred)
print(y_train.shape, y_train_pred.shape)

#actual and predicted Y of test data
print("Y test -> ", y_test)
print("Y test predicted -> ", y_test_pred)
print(y_test.shape, y_test_pred.shape)
```
