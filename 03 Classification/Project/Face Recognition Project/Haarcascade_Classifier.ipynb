# **Face/Object Detection**

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('kids2.jpeg')
plt.imshow(img)
```

```python
def plot(img):
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  plt.imshow(img)
  plt.axis(False)
  plt.figure(figsize=(8,8))
  plt.show()
  return img
```

```python
img = plot(img)
```

## **Haarcascade Classifier**
- object detection irrespective of scale in image and location
- fast
- can work well in real-time
- simple to implement
- less computing power required
- disadvatage : it predicts many false positives

```python
model = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
faces = model.detectMultiScale(img, scaleFactor=1.1, minNeighbors=1)
# 1.2 - scaling factor
# 2 - min neighbours
```

```python
faces
```

```python
faces.shape
# 4 faces with 4 parameters(x, y, height, width of the frame)
```

```python
def detectFace(img, faces):
  for f in faces:
    x, y, w, h = f
    cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,0), 2)
  plt.imshow(img)
  plt.show()
```

```python
detectFace(img, faces)
```
